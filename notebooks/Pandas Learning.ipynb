```python
df1= pd.DataFrame({'key': ['b','b','a','c','a','a','b'],'data1':range(7)})
```

```python
df2 = pd.DataFrame({'key': ['a','b','d'],'data2':range(3)})
```

```python
df2
```

```python
df1
```

```python
pd.merge(df1,df2) # Inner join on matching column
pd.merge(df1,df2, on='key') # Does the same as above
```

```python
pd.merge(df1,df2, on='key', how='left') # Left join 
```

```python
pd.merge(df1,df2, on='key', how='right') # RIght join 
```

```python
df3= pd.DataFrame({'lkey': ['b','b','a','c','a','a','b'],'data1':range(7)})
df4 = pd.DataFrame({'rkey': ['a','b','d'],'data2':range(3)})
```

```python
pd.merge(df3,df4, left_on='lkey',right_on='rkey') # Inner join on hetrogenious column
```

```python
pd.merge(df3,df4, left_on='lkey',right_on='rkey', how='outer') # outer join on hetrogenious column
```

```python
pd.merge(df3,df4, left_on='lkey',right_on='rkey', how='left', sort='lkey', left_index=True) # Left join on hetrogenious column
```

```python
pd.merge(df3,df4, left_on='lkey',right_on='rkey', how='left', sort='lkey', left_index=False) # Left join on hetrogenious column
```
