```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=9300,stop_words='english')
```

```python
bow=tfidf.fit_transform(cleaned_tweet)
```

```python
tfidf.get_feature_names()
```

```python
train_bow = bow[:38688,:]
test_bow = bow[38688:,:]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train, y_test = train_test_split(train_bow,label,test_size=0.2,random_state=3)
```

```python
from sklearn.naive_bayes import MultinomialNB
model= MultinomialNB(0.25)
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
metrics.accuracy_score(y_test,pred)
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
skf = StratifiedKFold(n_splits=10)
params = {'alpha': [0.001,0.01,0.1,0.2,0.5,0.7,0.9]}
nb = MultinomialNB()
gs = GridSearchCV(nb, cv=skf, param_grid=params, return_train_score=True)
```
